1.	Read the iris dataset using Pandas (ref Option 2 above)
2.	Extract the Iris-setosa and Iris-verginica samples only
3.	Extract the petal length and petal width. These shall be our features.

```python
import pandas as pd
raw_df = pd.read_csv('iris_data.csv', header=0)
raw_df
```

```python
flowerName = ['Iris-setosa','Iris-virginica']
#Filtering the dataset to only keep the rows which contains data of Iris Setorsa and Iris Virginica
df_set_ver = raw_df[raw_df[" class"].isin(flowerName)] 
df_set_ver
```

```python
# Filtering my dataframe to only keep the useful columns
df = df_set_ver[[" petal_length", " petal_width", " class"]]

# Split my dataframe into a trainig dataset and target values
X = df[[" petal_length", " petal_width"]]
y = df[[" class"]]

# Overview of the dataframes
df, X, y
```

4.	Plot the dataset as a scatter plot with the two Iris types shown as different colours

```python
#Ploting the data
colors = {'Iris-setosa': '#FF0000', 'Iris-virginica': '#0000FF'}
df.plot(kind='scatter', x=' petal_length', y=' petal_width', c=df[' class'].map(colors))
```

5.	Use the perceptron in sklearn.linear_model to train on the Iris data previously used. Set verbose = 1; have a look at the epochs, it converges very quickly?

```python
# So as not to confuse with our previously defined Perceptron class
from sklearn.linear_model import Perceptron as Perceptron

ppn = Perceptron(verbose = 1)

# Train our model with the trainig dataset and the target values
ppn.fit(X, y)
```

We can notice that from the Epoch 2, the loss already converge to 0. So the training converges very quickly. This is because the 2 class of flower can be very easily separated by linear hyperplane.

6.	Using many random inputs (e.g. 5000) plot to see if you can see the decision boundary for the feature scaled case

```python
import random

# Generating 5000 random inputs
# I noticed that petal_lenght is between 1.0 and 6.9 and petal_width is between 0.1 and 2.5 so I am generaring random values in theses ranges 
X_random = []
for x in range(5000):
  random.seed(x)
  X_random.append([1 + random.random() * 5.9,  0.1 + random.random() * 2.4])

# Overview of the random values
print(X_random)
```

```python
import pandas as pd

# Generating prediction baed on the random values dataset
y_pred = ppn.predict(X_random)

#Creating a dataframe with the random petal_width and petal_width and their corresponding class prediction
df_random = pd.DataFrame(X_random, columns = ['petal_length', 'petal_width'])
df_random['class'] = y_pred
display(df_random)


colors = {'Iris-setosa': '#FF0000', 'Iris-virginica': '#0000FF'}
df_random.plot(kind='scatter', x='petal_length', y='petal_width', c=df_random['class'].map(colors))
```

We can clearly see a decision boundary which split the blue dots (iris-virginica) and the red dots (iris-setosa).
