###Calculate and return the size of a matrix as a 2-dimensional tuple 

```python
#Check if the given parameter is a matrix (m,n) with m columns and n tuples of the same size
def checkIfMatrixCorrect(matrix):
  if type(matrix) != list:
    return False

  #Check if the first element is tuple and get its size
  wide = 0
  if type(matrix[0]) == list:
    wide = len(matrix[0])
  else:
    return False
  
  #check if all the tuples have the same size and only contrains number
  for elm in matrix:
    if type(elm) != list:
      return False
    if len(elm) != wide:
      return False
    for e in elm:
      if type(e) != float and type(e) != int:
        return False

  return True

print("[[2,2,1], [2,2]] -> matrix: ", checkIfMatrixCorrect([[2,2,1], [2,2]]))
print("[[2,2,1], [2,2,'2']] -> matrix: ", checkIfMatrixCorrect([[2,2,1], [2,2,"2"]]))
print("[[2,2,1], [2,[4],2]] -> matrix: ", checkIfMatrixCorrect([[2,2,1], [2,[4],2]]))
print("[[2,2,1], [2,4,3], [2,4,9]] -> matrix: ", checkIfMatrixCorrect([[2,2,1], [2,4,3], [2,4,9]]))
print("[[2,2,1]] -> matrix: ", checkIfMatrixCorrect([[2,2,1]]))
```

```python
def matrixSize(matrix):
  if checkIfMatrixCorrect(matrix) is False:
    raise ValueError('Parameter is not a matrix')
  l = len(matrix)
  m = len(matrix[0])
  return (l, m)

m1 = [[2,3,4],[1,2,3]]
print(matrixSize(m1))

#Test with a list which is not a matrix
m2 = [[2,3,4],[1,3]]
print(matrixSize(m2))
```

###Sum/subtract the matrix with another matrix of suitable size 

```python
def sumMatrix(m1, m2):
  size = matrixSize(m1)
  if size != matrixSize(m2):
    raise ValueError("Matrix don't have the same size")

  #initialize the matrix to return
  m3 = [ [0]*size[1] for i in range(size[0])]

  for indexTuple_, tuple_ in enumerate(m3):
    for indexNb, nb in enumerate(tuple_):
      m3[indexTuple_][indexNb] = m1[indexTuple_][indexNb] + m2[indexTuple_][indexNb]

  return m3

def substractMatrix(m1, m2):
  size = matrixSize(m1)
  if size != matrixSize(m2):
    raise ValueError("Matrix don't have the same size")

  #initialize the matrix to return
  m3 = [ [0]*size[1] for i in range(size[0])]

  for indexTuple_, tuple_ in enumerate(m3):
    for indexNb, nb in enumerate(tuple_):
      m3[indexTuple_][indexNb] = m1[indexTuple_][indexNb] - m2[indexTuple_][indexNb]

  return m3

m1 = [[2,3,4],[1,2,3]]
m2 = [[1,1,1],[5,1,1]]
print("m1:       [[2, 3, 4], [1 ,2 ,3]]")
print("m2:       [[1, 1, 1], [5, 1, 1]]")
print("Sum:     ", sumMatrix(m1,m2))
print("Substract", substractMatrix(m1,m2))
```

### Multiply the matrix with a matrix or vector of suitable size. 

```python
def multiplyMatrix(m1, m2): 
  sizeM1 = matrixSize(m1)
  sizeM2 = matrixSize(m2)
  if sizeM1[1] != sizeM2[0] :
    raise ValueError("Can't multiply these matrix. Sizes not conform")

  #initialize the matrix to return
  m3 = [ [0]*sizeM2[1] for i in range(sizeM1[0])]

  for y in range(sizeM1[0]):
    for i in range(sizeM2[1]):
      for j in range(sizeM2[0]):
        m3[y][i] += m1[y][j]*m2[j][i]
  
  return m3

m1 = [[2,3,4],[1,2,3]]
m2 = [[1,1],[5,1],[2,2]]
m3 = [[1],[5],[2]]
print("m1 * m2 = ", multiplyMatrix(m1, m2))
print("m1 * m3 = ", multiplyMatrix(m1, m3)) #Multiply a matrix with a vector 
print("m2 * m3 = ", multiplyMatrix(m2, m3)) #Try to multiply matrix with sizes which are not conform for a matrix multiplication 
```

###Calculate the determinant of a 2x2 matrix

```python
def determinant2x2(m1):
  if matrixSize(m1) != (2,2):
    raise ValueError("The matrix size is not 2x2")

  return m1[0][0] *  m1[1][1] - m1[1][0] * m1[0][1]

m1 = [[1, -4], [0, 3]]
print(determinant2x2(m1))
```

###Calculate the inverse of a 2x2 matrix

```python
def inverse2x2(m1):
  if matrixSize(m1) != (2,2):
    raise ValueError("The matrix size is not 2x2")
  det = determinant2x2(m1)
  if det == 0:
    raise ValueError("Matrix inversion is impossible -> determinant = 0")

  m0 = [[m1[1][1], -m1[0][1]],[-m1[1][0], m1[0][0]]]
  for i in range(2):
    for j in range(2):
      m0[i][j] = (1.0 / det ) * m0[i][j]
  
  return m0

m1 =  [[5, 2], [-7, -3]]
print("determinant = ", determinant2x2(m1))
print("inverse = ", inverse2x2(m1))

m2 = [[0, -4], [0, 3]] #Try to calcul the inverse of matrix with a determinant egal to 0
print("determinant = ", determinant2x2(m2))
print("inverse = ", inverse2x2(m2))
print(inverse2x2(m2))
```

###Calculate the transpose of an mxn matrix. 

```python
def transposeMatrix(matrix):
  size = matrixSize(matrix)
  x = size[1]
  y = size[0]
  m = [ [0]*size[0] for i in range(size[1])]

  for i in range(y):
    for j in range(x):
      m[j][i] = matrix[i][j]
  
  return m

m1 = [[1,1],[5,1],[2,2]]
print(transposeMatrix(m1)) 
```
